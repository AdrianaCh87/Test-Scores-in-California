# Standardized Tests Performance in California Schools

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.options.display.max_columns = 30
```

## Data Gathering

Dataset: 2019 SAT Scores in California by School

Source: California Department of Education https://www.cde.ca.gov/ds/sp/ai/

Description: The information in this dataset contains the results for 12th graders and 11th graders across all school districts in CA. The scores included are for two tests: ERW (English/Reading/Writing) and Math. The number of enrolled students in the high school and the number of test takers is also documented. This data has information at the school, district, and county level. 

```python
sat_file = 'sat_2019_ca.csv'
file_path = '../data/'
```

### SAT Dataset:

```python
sat = pd.read_csv(file_path + sat_file)
sat.head()  # Check the first five rows of the dataset
```

```python
sat.tail() # Check the last five rows of the dataset
```

Remove column "Unnamed: 25" and the last row of the dataset since they do not have any information

```python
sat.drop(columns = 'Unnamed: 25', inplace = True)
sat.drop([2579], inplace = True)
```

## Data Cleaning

Statistical summary of dataset:

```python
sat.describe(include='all')
```

Check for null values:

```python
sat.isnull().sum()
```

The missing values in column "SName" are expected because this field belongs to record type "S" for School, and there are 597 records for District and County which will not have an SName. In a similar way, the missing values in column "DName" are expected too because this field belongs to record type "D" for District, and there are 58 records for County which will not have a DName. Below is the breakdown of the record types in the dataset which coincide with the missing values seeing above.

```python
sat['RType'].value_counts()
```

The missing values in the score columns is because per the data dictionary if zero (0) students took the SAT at a school, district, or county, but grade 12 enrollment was equal to or greater than 15, then an ‘N/A’ is used instead of the number or percentage for privacy reasons.

Check the data type for each variable to make sure they are correct:

```python
sat.dtypes
```

Upon review of the variable data types, the columns with codes should be strings and not numerical. A function will be created to convert the data type of these columns

```python
# Convert the value to string 
def to_string(cell):
    return str(cell)
```

```python
sat['CDS'] = sat['CDS'].map(to_string)
sat['CCode'] = sat['CCode'].map(to_string)
sat['CDCode'] = sat['CDCode'].map(to_string)
sat['SCode'] = sat['SCode'].map(to_string)
```

For the columns containing a score, replace the asterisk (*) with NaN, so that the columns datatype is numerical

```python
def clean_score(cell):
    if cell == '*':
        return np.nan
    else:
        return float(cell)
```

```python
sat['NumERWBenchmark12'] = sat['NumERWBenchmark12'].map(clean_score)
sat['PctERWBenchmark12'] = sat['PctERWBenchmark12'].map(clean_score)
sat['NumMathBenchmark12'] = sat['NumMathBenchmark12'].map(clean_score)
sat['PctMathBenchmark12'] = sat['PctMathBenchmark12'].map(clean_score)
sat['NumERWBenchmark11'] = sat['NumERWBenchmark11'].map(clean_score)
sat['PctERWBenchmark11'] = sat['PctERWBenchmark11'].map(clean_score)
sat['NumMathBenchmark11'] = sat['NumMathBenchmark11'].map(clean_score)
sat['PctMathBenchmark11'] = sat['PctMathBenchmark11'].map(clean_score)
sat['TotNumBothBenchmark12'] = sat['TotNumBothBenchmark12'].map(clean_score)
sat['PctBothBenchmark12'] = sat['PctBothBenchmark12'].map(clean_score)
sat['TotNumBothBenchmark11'] = sat['TotNumBothBenchmark11'].map(clean_score)
sat['PctBothBenchmark11'] = sat['PctBothBenchmark11'].map(clean_score)
```

Rename the column names to lower and camel case for easier reference:

```python
{c: c.lower() for c in sat.columns}

sat_new_columns_dict = {
 'CDS': 'code',
 'CCode': 'county_code',
 'CDCode': 'district_code',
 'SCode': 'school_code',
 'RType': 'record_type',
 'SName': 'school_name',
 'DName': 'district_name',
 'CName': 'county_name',
 'Enroll12': '12grade_enrollment',
 'NumTSTTakr12': '12grade_test_takers',
 'NumERWBenchmark12': '12graders_passing_ERW',
 'PctERWBenchmark12': '12graders%_passing_ERW',
 'NumMathBenchmark12': '12graders_passing_Math',
 'PctMathBenchmark12': '12graders%_passing_Math',
 'Enroll11': '11grade_enrollment',
 'NumTSTTakr11': '11grade_test_takers',
 'NumERWBenchmark11': '11graders_passing_ERW',
 'PctERWBenchmark11': '11graders%_passing_ERW',
 'NumMathBenchmark11': '11graders_passing_Math',
 'PctMathBenchmark11': '11graders%_passing_Math',
 'TotNumBothBenchmark12': '12graders_passing_both_tst',
 'PctBothBenchmark12': '12graders%_passing_both_tst',
 'TotNumBothBenchmark11': '11graders_passing_both_tst',
 'PctBothBenchmark11': '11graders%_passing_both_tst',
 'Year': 'test_year'}

sat.rename(columns=sat_new_columns_dict, inplace = True)
```

## Exploratory Data Analysis

### Create new features

Create new columns for sum the number of 11th graders and 12th graders to have a total amount of students

```python
sat['all_test_takers'] = sat['12grade_test_takers'] + sat['11grade_test_takers']
sat['all_enrollment'] = sat['12grade_enrollment'] + sat['11grade_enrollment']
sat['stdts_passing_both_tst'] = sat['12graders_passing_both_tst'] + sat['11graders_passing_both_tst'] 
sat['stdts_passing_math_tst'] = sat['12graders_passing_Math'] + sat['11graders_passing_Math'] 
sat['stdts_passing_erw_tst'] = sat['12graders_passing_ERW'] + sat['11graders_passing_ERW'] 
```

```python
counties = sat['county_name'].nunique()
print(f'There are {counties} counties in this dataset')
districts = sat['district_name'].nunique()
print(f'There are {districts} school districts in this dataset')
schools = sat['school_name'].nunique()
print(f'There are {schools} schools in this dataset')
```

Evaluate the size of each county in terms of number of schools:

```python
plt.figure(figsize=(14,12))
sat.groupby('county_name')['school_name'].count().sort_values(ascending=False).plot.barh()
plt.title('Number of schools in each county', size=15)
plt.xlabel('Number of schools', size=12)
plt.ylabel('Counties in California', size=12);
```

Evaluate the number of total students taking the SAT in each county:

```python
test_takers_county = sat.groupby('county_name')['all_test_takers'].sum()
pct_test_takers_county = test_takers_county/sat.groupby('county_name')['all_enrollment'].sum()*100
pct_test_takers_county = pct_test_takers_county.sort_values(ascending=False)

plt.figure(figsize=(14,12))
pct_test_takers_county.plot.barh()
plt.title('SAT Test Takers in each County', size=15)
plt.xlabel('Percent of Test Takers out of Total Students Enrolled in School', size=12)
plt.ylabel('Counties in California', size=12);
```

Explore student SAT scores in each county:

```python
students_passing_by_county = sat.groupby('county_name')['stdts_passing_both_tst'].sum()
pct_passing_county = students_passing_by_county/sat.groupby('county_name')['all_test_takers'].sum()*100
pct_passing_county = pct_passing_county.sort_values(ascending=False)

plt.figure(figsize=(14,12))
pct_passing_county.plot.barh()
plt.title('Students Meeting the Benchmark Scores in both SAT tests', size=15)
plt.xlabel('Percent of students out of all Test Takers', size=12)
plt.ylabel('Counties in California', size=12);
```

Number of Test Takers in 11th and 12th grade:

```python
plt.figure(figsize=(8,5))
sat[['11grade_test_takers','12grade_test_takers']].sum().plot.bar()
plt.title('Students taking the SAT in the 11th and 12th grade in California', size=15)
plt.ylabel('Number of Students', size=12)
plt.xticks(rotation=0);
```

Difference in test scores between 11th graders and 12th graders:

```python
students_by_county = sat.groupby('county_name')[['11graders_passing_both_tst',
                                                 '12graders_passing_both_tst',
                                                '11grade_test_takers',
                                                '12grade_test_takers']].sum()
students_by_county['11graders_met_benchmark'] = (students_by_county['11graders_passing_both_tst']/students_by_county['11grade_test_takers'])*100
students_by_county['12graders_met_benchmark'] = (students_by_county['12graders_passing_both_tst']/students_by_county['12grade_test_takers'])*100

# Set up ticks and their labels for the graph
ticks=list(np.arange(0, len(students_by_county.index), 1))
labels=list(students_by_county.index)

students_by_county[['11graders_met_benchmark','12graders_met_benchmark']].plot(figsize=(14,8));
plt.title('Percent of Students in each Grade Meeting the Benchmark SAT Scores')
plt.xlabel('Counties in California', size=12)
plt.xticks(ticks=ticks, labels=labels, rotation=90)
plt.ylabel('% of Students out of all Test Takers who met Bechmark Scores', size=12);
```

Explore the correlation between math SAT scores and English/Reading/Writing scores:

```python
df_12_passing = sat[sat['record_type']=='S'][['12graders_passing_Math',
                                              '12graders_passing_ERW', 
                                              '11graders_passing_Math',
                                              '11graders_passing_ERW']]
df_12_passing.dropna(inplace=True)

plt.figure(figsize=(8,8))
plt.scatter(df_12_passing['12graders_passing_ERW'], df_12_passing['12graders_passing_Math'],label='12th Grade Students', marker='.')
plt.scatter(df_12_passing['11graders_passing_ERW'], df_12_passing['11graders_passing_Math'],label='11th Grade Students', marker='1')
plt.title('Number of students meeting the ERW and Math test scores', size=15)
plt.xlabel('Students meeting the Evidence-Based Reading & Writing (ERW) benchmark')
plt.ylabel('Students meeting the Math benchmark')
plt.legend();
```
